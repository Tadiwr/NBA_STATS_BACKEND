Getting Offensive Stats

```python
import requests as req
import json

url = "http://sports.core.api.espn.com/v2/sports/basketball/leagues/nba/seasons/2023/types/2/teams/9/statistics"
stats_json = json.loads(req.get(url=url).text)

stats_json.keys()
```

```python
off_stats = stats_json['splits']["categories"][2]["stats"]
off_stats
```

Writing Offensive stats into a csv file

```python
# stats_json into a csv file
import pandas as pd
off_stats = pd.DataFrame(off_stats)
off_stats.to_csv('off_stats.csv', index=False)
```
